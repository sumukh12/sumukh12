```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.amazon.in/Smartphones-%E2%82%B910-000-%E2%82%B920-Basic-Mobiles/s?rh=n%3A1805560031%2Cp_36%3A1318506031')
```

```python
page
```

```python
#loading webpage source code
page.content
```

```python
#loading html.parser
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
Mobile=soup.find_all("span" ,class_="a-size-base-plus a-color-base a-text-normal")
```

```python
Mobile
```

```python
Mobile_list=[]
for i in Mobile:
    Mobile_list.append(i.text.replace('\n',''))
Mobile_list
```

```python
Price=soup.find_all('span',class_='a-price-whole')
```

```python
Price
```

```python
Price_list=[]
for i in Price:
    Price_list.append(i.text.replace('\n',''))
Price_list
```

```python
Image=soup.find_all('div', class_="a-section aok-relative s-image-square-aspect")
```

```python
Image
```

```python
Image_url=[]
for i in Image:
    Image_url.append(i.text.replace('\n',''))
Image_url
```

```python
Rating=soup.find_all('i', class_="a-icon a-icon-star-small a-star-small-5 aok-align-bottom")
```

```python
Rating
```

```python
Rat_list=[]
for i in Rating:
    Rat_list.append(i.text.replace('\n',''))
Rat_list
```
