```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,"html.parser")
```

```python
soup
```

```python
titles=soup.find_all("h1",class_="firstHeading")
```

```python
titles
```

```python
web_titles=[]
for i in titles:
    web_titles.append(i.text.replace('\n',''))
web_titles
```

```python
Today_Feature_article=soup.find_all('h2',class_="mp-h2")
```

```python
Today_Feature_article
```

```python
Feature_article=[]
for i in Today_Feature_article:
    Feature_article.append(i.text.replace('\n',''))
Feature_article
```

```python
titles=soup.find_all(["h1","h2"])
print("List of all the header tag:",*titles,sep="\n\n")
```
