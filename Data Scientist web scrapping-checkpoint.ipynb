```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.monster.com/jobs/search/?q=Data-Scientist&where=New-Delhi')
page
```

```python
#loading webpage source code
page.content
```

```python
#loading html parser
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
job_title=soup.find_all('h2',class_='title')
job_title
```

```python
job=[]
for i in job_title:
    job.append(i.text.replace('/n',''))
job
```

```python
print(len(job))
```

```python
import pandas as pd
```

```python
Data_Scientist=pd.DataFrame({})
Data_Scientist['job_title']=job[:26]
print(Data_Scientist)
```
