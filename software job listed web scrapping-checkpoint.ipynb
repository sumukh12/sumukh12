```python
#importing required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.monster.com/jobs/q-software-developer-jobs')
page
```

```python
#webpage source code
page.content
```

```python
#html parser
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
job=soup.find_all('h2',class_='title')
```

```python
job
```

```python
jobs=[]
for i in job:
    jobs.append(i.text.replace('\n',''))
jobs
```

```python
print(len(jobs))
```

```python
import pandas as pd
```

```python
job_listed=pd.DataFrame({})
job_listed["job"]=jobs[:20]
```
